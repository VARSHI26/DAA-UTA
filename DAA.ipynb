```python
import random
import time
import matplotlib.pyplot as plt
```

### This block of code is used to generate random number from 1 to 100

```python
def make_random_nums(n, low=1, high=100):
    nums = []
    for i in range(n):
        nums.append(random.randint(low, high))
    return nums
```

### Generating different data sets

```python
num_sizes = [5, 10, 20, 30, 40, 50, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 2000, 3000, 4000, 5000]
data = {}

for s in num_sizes:
    data[s] = make_random_nums(s)
```

###### Insertion Sort is a simple sorting algorithm that works by dividing the array into a sorted and an unsorted portion. It starts with the first element as already sorted. Then, for each element in the unsorted portion, it finds the correct position in the sorted part by comparing it with elements one by one. The element is then inserted at its correct position, and the process continues until the entire array is sorted. This method is similar to sorting playing cards in your hand, where you place each card in the right spot relative to the already sorted cards.

```python
def insertion_sort(lst):
    for x in range(1, len(lst)):
        temp = lst[x]
        y = x - 1
        while y >= 0:
            if lst[y] > temp:
                lst[y + 1] = lst[y]
                y -= 1
            else:
                break
        lst[y + 1] = temp
```

###### Selection Sort is a straightforward sorting algorithm that works by repeatedly selecting the smallest element from the unsorted portion of the array and swapping it with the first unsorted element. The process starts with the entire array as unsorted, and the algorithm gradually builds the sorted portion from left to right. In each step, the smallest element from the unsorted part is found and placed in its correct position. This process continues until the entire array is sorted. It’s like organizing a group of items by repeatedly picking the smallest item and moving it to the correct spot

```python
def selection_sort(lst):
    size = len(lst)
    for a in range(size):
        small = a
        for b in range(a + 1, size):
            if lst[b] < lst[small]:
                small = b
        temp = lst[a]
        lst[a] = lst[small]
        lst[small] = temp
```

###### Bubble Sort is a simple sorting algorithm that repeatedly steps through the array, compares adjacent elements, and swaps them if they are in the wrong order. This process is repeated for all elements until no swaps are needed, indicating that the array is sorted. During each pass, the largest unsorted element “bubbles up” to its correct position at the end of the array. This method continues, shrinking the unsorted portion of the array with each pass. It’s similar to sorting items by repeatedly comparing and swapping neighboring items until everything is in order.

```python
def bubble_sort(lst):
    size = len(lst)
    for a in range(size):
        did_swap = False
        for b in range(0, size - a - 1):
            if lst[b] > lst[b + 1]:
                temp = lst[b]
                lst[b] = lst[b + 1]
                lst[b + 1] = temp
                did_swap = True
        if did_swap == False:
            break
```

```python
def measure_time(sort_function, data):
    start = time.time()
    sort_function(list(data))
    return time.time() - start
```

```python
adjusted_runtimes = {
    "Insertion Sort": [],
    "Selection Sort": [],
    "Bubble Sort": []
}

for size in num_sizes:
    for sort_name, sort_function in [("Insertion Sort", insertion_sort),
                                     ("Selection Sort", selection_sort),
                                     ("Bubble Sort", bubble_sort)]:
        runtime = measure_time(sort_function, data[size])
        adjusted_runtimes[sort_name].append(runtime)
```

```python
for sort_name, times in adjusted_runtimes.items():
        plt.plot(num_sizes, times, marker='o', label=sort_name)

plt.title("Sorting Algorithm Runtimes")
plt.xlabel("Input Size (n)")
plt.ylabel("Runtime (seconds)")
plt.legend()
plt.grid(True)
plt.show()
```
